```
import nltk
```

Word Tokenization

```
nltk.download('punkt')
```

```
from nltk.tokenize import word_tokenize
input = "Have a nice day"
print(word_tokenize(input))
```

Sentence Tokenization

```
from nltk.tokenize import sent_tokenize
input = "Taj Mahal is in Agra. One of the seven wonders of the world."
print(sent_tokenize(input))
```

Tokenize using Regular Expression

```
from nltk.tokenize import RegexpTokenizer 
t = RegexpTokenizer('\s+', gaps = True)
input = "Maybe I am the happiest person"
tokens = t.tokenize(input) 
print(tokens) 
```

Count word frequency

```
input = "The most wellknown and full NLP Library which is free and open source having many thirdparty extensions it has plenty of approaches to each NLP task"
tokens = word_tokenize(input)
f = nltk.FreqDist(tokens)
for key,val in f.items():
    print (str(key) + ':' + str(val))
```

Plot a graph for these tokens:

```
f.plot(24, cumulative=False)
```

Importing StopWords

```
nltk.download('stopwords')
```

```
from nltk.corpus import stopwords
stop_words = stopwords.words('english')
```

Remove stopwords from given tokens

```
input = "The most well-known and full NLP Library which is free and open source Having many third-party extensions it has plenty of approaches to each NLP task"
tokens = word_tokenize(input)
for i in tokens:
    if i in stop_words:
        tokens.remove(i)
print("After removing Stop Words: ",tokens)
```

Get definitions and examples for a given word using WordNet

```
from nltk.corpus import wordnet
s = wordnet.synsets("dream")
print(s[0].definition())
print(s[0].examples())
```

Get synonyms for a given word using WordNet

```
from nltk.corpus import wordnet
synonyms_list = []
for i in wordnet.synsets('dream'):
    for j in i.lemmas():
        synonyms_list.append(j.name())
print(synonyms_list)
```

Get antonyms using WordNet

```
from nltk.corpus import wordnet
antonyms_list = []
for i in wordnet.synsets("down"):
    for j in i.lemmas():
        if j.antonyms():
            antonyms_list.append(j.antonyms()[0].name())
print(antonyms_list)
```

POS TAGGING

```
nltk.download('averaged_perceptron_tagger')
```

```
from nltk import pos_tag
print("Before POS Tagging:",tokens)
tagged_tokens = pos_tag(tokens)
print("After POS Tagging:",tagged_tokens)
```

Stemming

```
from nltk.stem import PorterStemmer

input = ["laugh","laughs","laughed","laughing"]
p = PorterStemmer()

for i in input:
    print( i , " : ",p.stem(i))
```

Lemmatization

```
nltk.download('wordnet')
```

```
from nltk.stem import WordNetLemmatizer
wordnet_lemma = WordNetLemmatizer()
input = "dries dry drives driving"
tokens = word_tokenize(input)
for i in tokens:
    print("Lemma for {} is {}".format(i,wordnet_lemma.lemmatize(i)))
```
