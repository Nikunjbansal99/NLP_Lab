**Read it as a .csv file and create a dataframe.**

```
import pandas as pd
dataset = pd.read_csv('train_triplets_sample.txt',nrows = 10000, sep = '\t', names = ['User','Song','Play_Count'], header=0)
```

**Display it.**

```
dataset.head(10)
```

**Display per unique user play_count.**

```
User_List = []
Play_Count_List = []
for ind in dataset.index: 
    u = dataset['User'][ind]
    p = dataset['Play_Count'][ind]
    if u in User_List:
        value = User_List.index(u)
        Play_Count_List[value] += p
    else:
        User_List.append(u)
        Play_Count_List.append(p)
```

```
df_task2 = pd.DataFrame(list(zip(User_List, Play_Count_List)), columns =['User', 'Play_Count']) 
df_task2.head(32)
```

**Display per unique song play_count.**

```
Song_List = []
Play_Count_List = []
for ind in dataset.index: 
    s = dataset['Song'][ind]
    p = dataset['Play_Count'][ind]
    if s in Song_List:
        value = Song_List.index(s)
        Play_Count_List[value] += p
    else:
        Song_List.append(s)
        Play_Count_List.append(p)
```

```
df_task3 = pd.DataFrame(list(zip(Song_List, Play_Count_List)), columns =['Song', 'Play_Count']) 
df_task3.head(21)
```

**Display top users with respect to play_count and save as different .csv files.**

```
top_users_df = df_task2.sort_values(by=['Play_Count'], ascending=False)
top_users_df.to_csv('top_users.csv',index=False)
```



**Display top songs with respect to play_count and save as different .csv files.**

```
top_songs_df = df_task3.sort_values(by=['Play_Count'], ascending=False)
top_songs_df.to_csv('top_songs.csv',index=False)
```
